```python
# Convert the data to prompt-response format
prompt_response_data = []
for entry in data_fin:
    prompt = f"I'm a {entry['age']}-year-old {entry['gender']} looking to invest in {entry['Avenue']} for {entry['Purpose']} over the next {entry['Duration']}. What are my options?"
    response = (
        f"Based on your preferences, here are your investment options:\n"
        f"- Mutual Funds: {entry['Mutual_Funds']}\n"
        f"- Equity Market: {entry['Equity_Market']}\n"
        f"- Debentures: {entry['Debentures']}\n"
        f"- Government Bonds: {entry['Government_Bonds']}\n"
        f"- Fixed Deposits: {entry['Fixed_Deposits']}\n"
        f"- PPF: {entry['PPF']}\n"
        f"- Gold: {entry['Gold']}\n"
        f"Factors considered: {entry['Factor']}\n"
        f"Objective: {entry['Objective']}\n"
        f"Expected returns: {entry['Expect']}\n"
        f"Investment monitoring: {entry['Invest_Monitor']}\n"
        f"Reasons for choices:\n"
        f"- Equity: {entry['Reason_Equity']}\n"
        f"- Mutual Funds: {entry['Reason_Mutual']}\n"
        f"- Bonds: {entry['Reason_Bonds']}\n"
        f"- Fixed Deposits: {entry['Reason_FD']}\n"
        f"Source of information: {entry['Source']}\n"
    )
    prompt_response_data.append({"prompt": prompt, "response": response})

prompt_response_data[:5]
```

```python
from langchain.docstore.document import Document
documents = []
for entry in prompt_response_data:
    combined_text = f"Prompt: {entry['prompt']}\nResponse: {entry['response']}"
    documents.append(Document(page_content=combined_text))
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=10)
texts = text_splitter.split_documents(documents)
```

```python
from typing import List, Dict
from langchain.chains import RetrievalQA

# ... (Your existing LangChain setup, loading data, etc.) ...

def calculate_relevancy(response: str, query: str) -> bool:
    """Determines if a response is relevant to a query.

    Args:
        response: The generated response from the QA system.
        query: The user's query.

    Returns:
        True if the response is relevant, False otherwise.
    """
    # Implement your relevancy logic here
    # Example: Check if the query terms are present in the response
    query_terms = query.lower().split()
    response_lower = response.lower()

    # Consider response relevant if at least one query term is present
    is_relevant = any(term in response_lower for term in query_terms)
    return is_relevant

def calculate_precision(qa_chain: RetrievalQA, queries: List[str]) -> float:
    """Calculates the precision of a question-answering system.

    Args:
        qa_chain: The RetrievalQA chain object representing the QA system.
        queries: A list of user queries.

    Returns:
        The precision as a float between 0 and 1.
    """
    relevant_responses = 0
    total_responses = 0

    for query in queries:
        result = qa_chain({"query": query})
        generated_answer = result['result']

        if generated_answer:  # Check if there is a response
            total_responses += 1
            if calculate_relevancy(generated_answer, query):
                relevant_responses += 1

    if total_responses == 0:
        return 0  # Avoid division by zero if there are no responses
    else:
        precision = relevant_responses / total_responses
        return precision


# Example Usage:
queries = [
    "I'm a 20-year-old female looking to invest in Mutual Fund for Wealth Creation over the next 15 years. What are my options?",
    "What are the risks of investing in penny stocks?",  # Add more queries here
    # ...
]

# Assuming `qa` is your RetrievalQA object set up previously
precision = calculate_precision(qa, queries)
print(f"Precision: {precision}")
```
